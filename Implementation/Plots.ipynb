# **Import the Necessary Libraries**

```python
import matplotlib.pyplot as plt
import numpy as np

models = ['RNN', 'LSTM', 'GRU']

# Values from your table
train_mae_urban = [2.7975, 2.7649, 2.7536]
test_mae_urban  = [1.9773, 1.9643, 1.9447]
train_mae_highway = [1.6083, 0.6094, 0.7292]
test_mae_highway  = [0.6139, 0.3733, 0.3954]

x = np.arange(len(models))
width = 0.35

fig, axs = plt.subplots(2, 1, figsize=(8, 10))

# Urban subplot
rects1 = axs[0].bar(x - width/2, train_mae_urban, width, label='Train MAE')
rects2 = axs[0].bar(x + width/2, test_mae_urban, width, label='Test MAE')

axs[0].set_ylabel('MAE')
axs[0].set_title('Urban Scenario MAE for RNN, LSTM, and GRU')
axs[0].set_xticks(x)
axs[0].set_xticklabels(models)
axs[0].legend()

for rect in rects1 + rects2:
    height = rect.get_height()
    axs[0].annotate(f'{height:.4f}',
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),
                    textcoords='offset points',
                    ha='center', va='bottom')

# Highway subplot
rects3 = axs[1].bar(x - width/2, train_mae_highway, width, label='Train MAE')
rects4 = axs[1].bar(x + width/2, test_mae_highway, width, label='Test MAE')

axs[1].set_ylabel('MAE')
axs[1].set_title('Highway Scenario MAE for RNN, LSTM, and GRU')
axs[1].set_xticks(x)
axs[1].set_xticklabels(models)
axs[1].legend()

for rect in rects3 + rects4:
    height = rect.get_height()
    axs[1].annotate(f'{height:.4f}',
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),
                    textcoords='offset points',
                    ha='center', va='bottom')

plt.xlabel('Models')
plt.tight_layout()
plt.savefig('mae_comparison.pdf')
plt.show()
```
